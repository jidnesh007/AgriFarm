```python
# ============================================================
# COMPLETE CROP HEALTH ANALYSIS PIPELINE
# YOLOv8 + Disease Classification + Growth Stage + Recommendations
# ============================================================

import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch
from ultralytics import YOLO
from pathlib import Path
import json

print("âœ… All packages installed!")

# ============================================================
# 2ï¸âƒ£ DOWNLOAD DATASET FROM ROBOFLOW
# ============================================================

from roboflow import Roboflow

rf = Roboflow(api_key="oDrZn6mmOj8OOKnX8XXt")
project = rf.workspace("jidnesh-veunq").project("plants-diseases-detection-and-classification-w6gj6")
version = project.version(1)
dataset = version.download("yolov8")

print(f"âœ… Dataset downloaded to: {dataset.location}")

# ============================================================
# 3ï¸âƒ£ TRAIN YOLOV8 MODEL FOR DISEASE DETECTION
# ============================================================

# Train YOLOv8 Nano (lightweight for edge deployment)
model = YOLO('yolov8n.pt')  # Use yolov8s.pt for better accuracy

# Train the model
results = model.train(
    data=f'{dataset.location}/data.yaml',
    epochs=50,  # Increase to 100+ for production
    imgsz=640,
    batch=16,
    device='0',  # Use GPU
    patience=10,
    save=True,
    project='crop_disease_detection',
    name='yolov8_plant_disease',
    verbose=True
)

print("âœ… Training completed!")

# Save best model
best_model_path = 'crop_disease_detection/yolov8_plant_disease/weights/best.pt'
model = YOLO(best_model_path)

# ============================================================
# 4ï¸âƒ£ GROWTH STAGE ESTIMATION (Rule-based)
# ============================================================

def estimate_growth_stage(leaf_area_ratio, disease_severity):
    """
    Estimate growth stage based on visual cues
    leaf_area_ratio: normalized size of detected leaf (0-1)
    disease_severity: severity of disease (0-1)
    """
    if leaf_area_ratio < 0.3:
        stage = "Early/Seedling"
        stage_code = 1
    elif leaf_area_ratio < 0.6:
        stage = "Vegetative"
        stage_code = 2
    else:
        stage = "Flowering/Mature"
        stage_code = 3

    # Adjust based on disease severity
    if disease_severity > 0.7:
        confidence = 0.6  # Lower confidence if heavily diseased
    else:
        confidence = 0.85

    return {
        "stage": stage,
        "stage_code": stage_code,
        "confidence": confidence
    }

# ============================================================
# 5ï¸âƒ£ CROP TYPE MAPPING
# ============================================================

CROP_TYPE_MAPPING = {
    # Map your dataset classes to crop types
    'apple': 'Apple',
    'tomato': 'Tomato',
    'potato': 'Potato',
    'corn': 'Corn',
    'grape': 'Grape',
    'pepper': 'Pepper',
    'strawberry': 'Strawberry',
    'cherry': 'Cherry',
    'peach': 'Peach',
    'soybean': 'Soybean'
}

def extract_crop_type(class_name):
    """Extract crop type from class name"""
    class_lower = class_name.lower()
    for crop in CROP_TYPE_MAPPING.keys():
        if crop in class_lower:
            return CROP_TYPE_MAPPING[crop]
    return "Unknown"

# ============================================================
# 6ï¸âƒ£ DISEASE SEVERITY ESTIMATION
# ============================================================

def estimate_disease_severity(image, bbox):
    """
    Estimate disease severity based on color analysis
    """
    x1, y1, x2, y2 = map(int, bbox)
    roi = image[y1:y2, x1:x2]

    # Convert to HSV for color analysis
    hsv = cv2.cvtColor(roi, cv2.COLOR_BGR2HSV)

    # Calculate percentage of brown/yellow (disease indicators)
    # Brown: H=10-30, S=100-255, V=20-200
    lower_brown = np.array([10, 100, 20])
    upper_brown = np.array([30, 255, 200])
    brown_mask = cv2.inRange(hsv, lower_brown, upper_brown)

    # Yellow: H=20-40, S=100-255, V=100-255
    lower_yellow = np.array([20, 100, 100])
    upper_yellow = np.array([40, 255, 255])
    yellow_mask = cv2.inRange(hsv, lower_yellow, upper_yellow)

    disease_pixels = cv2.countNonZero(brown_mask) + cv2.countNonZero(yellow_mask)
    total_pixels = roi.shape[0] * roi.shape[1]

    severity_ratio = disease_pixels / total_pixels if total_pixels > 0 else 0

    if severity_ratio < 0.1:
        severity = "Low"
        severity_score = 1
    elif severity_ratio < 0.3:
        severity = "Moderate"
        severity_score = 2
    else:
        severity = "High"
        severity_score = 3

    return {
        "severity": severity,
        "severity_score": severity_score,
        "affected_percentage": round(severity_ratio * 100, 2)
    }

# ============================================================
# 7ï¸âƒ£ KNOWLEDGE BASE FOR RECOMMENDATIONS
# ============================================================

DISEASE_RECOMMENDATIONS = {
    "apple_scab": {
        "treatment": "Apply sulfur-based fungicide",
        "preventive": "Improve air circulation, remove infected leaves",
        "organic": "Neem oil spray weekly",
        "timing": "Early morning application"
    },
    "tomato_early_blight": {
        "treatment": "Apply copper-based fungicide",
        "preventive": "Mulch around plants, water at base",
        "organic": "Baking soda solution (1 tbsp per gallon)",
        "timing": "Every 7-10 days"
    },
    "potato_late_blight": {
        "treatment": "Apply mancozeb or chlorothalonil",
        "preventive": "Proper spacing, avoid overhead irrigation",
        "organic": "Copper sulfate spray",
        "timing": "Before rain or high humidity"
    },
    "corn_rust": {
        "treatment": "Apply triazole fungicide",
        "preventive": "Plant resistant varieties, crop rotation",
        "organic": "Sulfur dust application",
        "timing": "At first sign of disease"
    },
    "grape_black_rot": {
        "treatment": "Apply myclobutanil fungicide",
        "preventive": "Prune for air flow, remove mummified fruit",
        "organic": "Bordeaux mixture spray",
        "timing": "Pre-bloom to post-harvest"
    },
    "healthy": {
        "treatment": "No treatment needed",
        "preventive": "Continue regular monitoring and good practices",
        "organic": "Maintain soil health with compost",
        "timing": "Regular inspection weekly"
    },
    "default": {
        "treatment": "Consult local agricultural extension office",
        "preventive": "Monitor crop regularly, maintain plant health",
        "organic": "Use organic matter, practice crop rotation",
        "timing": "As needed based on symptoms"
    }
}

def get_recommendation(crop_type, disease_name, severity_score, growth_stage_code):
    """
    Generate deterministic recommendation based on diagnosis
    """
    # Normalize disease name for lookup
    disease_key = disease_name.lower().replace(" ", "_").replace("-", "_")

    # Get base recommendation
    if disease_key in DISEASE_RECOMMENDATIONS:
        rec = DISEASE_RECOMMENDATIONS[disease_key]
    else:
        rec = DISEASE_RECOMMENDATIONS["default"]

    # Adjust based on severity
    if severity_score == 3:  # High severity
        urgency = "URGENT"
        action = "Immediate treatment required"
    elif severity_score == 2:  # Moderate
        urgency = "MODERATE"
        action = "Treat within 3-5 days"
    else:  # Low
        urgency = "LOW"
        action = "Monitor and apply preventive measures"

    # Adjust based on growth stage
    if growth_stage_code == 1:  # Early
        note = "Young plants are vulnerable - act quickly"
    elif growth_stage_code == 2:  # Vegetative
        note = "Critical growth phase - protect foliage"
    else:  # Flowering/Mature
        note = "Protect yield - avoid harsh chemicals near harvest"

    return {
        "urgency": urgency,
        "action": action,
        "treatment": rec["treatment"],
        "preventive": rec["preventive"],
        "organic_option": rec["organic"],
        "timing": rec["timing"],
        "growth_note": note
    }

# ============================================================
# 8ï¸âƒ£ FARMER-FRIENDLY EXPLANATION GENERATOR
# ============================================================

def generate_farmer_explanation(diagnosis, recommendation):
    """
    Convert technical diagnosis to simple language
    """
    crop = diagnosis['crop_type']
    disease = diagnosis['disease_name']
    severity = diagnosis['severity']['severity']
    stage = diagnosis['growth_stage']['stage']

    # Simple explanation
    if disease.lower() == "healthy":
        explanation = f"""
ğŸŒ± GOOD NEWS! Your {crop} looks healthy.

âœ… What we found:
   - No disease detected
   - Plant is in {stage} stage
   - Continue your current care routine

ğŸ“‹ What to do:
   - Keep monitoring your crop weekly
   - Maintain good watering and fertilization
   - Watch for any changes in leaf color or spots
"""
    else:
        explanation = f"""
âš ï¸ ATTENTION NEEDED for your {crop}

ğŸ” What we found:
   - Disease: {disease}
   - Severity: {severity}
   - Growth Stage: {stage}
   - Affected Area: {diagnosis['severity']['affected_percentage']}%

ğŸ¯ Urgency Level: {recommendation['urgency']}

ğŸ“‹ What to do NOW:
   1. {recommendation['action']}
   2. Treatment: {recommendation['treatment']}
   3. When: {recommendation['timing']}

ğŸŒ¿ Natural Option:
   - {recommendation['organic_option']}

ğŸ’¡ Prevention for Future:
   - {recommendation['preventive']}

â° Important Note:
   - {recommendation['growth_note']}
"""

    return explanation.strip()

# ============================================================
# 9ï¸âƒ£ COMPLETE INFERENCE PIPELINE
# ============================================================

def analyze_crop_image(image_path, model, confidence_threshold=0.5):
    """
    Complete pipeline for crop health analysis
    """
    # Load image
    image = cv2.imread(image_path)
    if image is None:
        return {"error": "Could not load image"}

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    h, w = image.shape[:2]

    # Run YOLOv8 detection
    results = model.predict(image_path, conf=confidence_threshold, verbose=False)

    if len(results[0].boxes) == 0:
        return {"error": "No crop or disease detected in image"}

    # Get best detection (highest confidence)
    boxes = results[0].boxes
    best_idx = boxes.conf.argmax()
    best_box = boxes[best_idx]

    # Extract information
    class_id = int(best_box.cls[0])
    class_name = results[0].names[class_id]
    confidence = float(best_box.conf[0])
    bbox = best_box.xyxy[0].cpu().numpy()

    # Extract crop type and disease
    crop_type = extract_crop_type(class_name)
    disease_name = class_name if "healthy" not in class_name.lower() else "Healthy"

    # Calculate leaf area ratio
    box_area = (bbox[2] - bbox[0]) * (bbox[3] - bbox[1])
    image_area = h * w
    leaf_area_ratio = box_area / image_area

    # Estimate disease severity
    severity_info = estimate_disease_severity(image, bbox)

    # Estimate growth stage
    growth_info = estimate_growth_stage(leaf_area_ratio, severity_info['severity_score'] / 3)

    # Build diagnosis
    diagnosis = {
        "crop_type": crop_type,
        "disease_name": disease_name,
        "confidence": round(confidence, 3),
        "severity": severity_info,
        "growth_stage": growth_info,
        "bbox": bbox.tolist()
    }

    # Generate recommendation
    recommendation = get_recommendation(
        crop_type,
        disease_name,
        severity_info['severity_score'],
        growth_info['stage_code']
    )

    # Generate farmer-friendly explanation
    explanation = generate_farmer_explanation(diagnosis, recommendation)

    # Visualize results
    annotated_image = visualize_results(image_rgb, diagnosis, recommendation)

    return {
        "diagnosis": diagnosis,
        "recommendation": recommendation,
        "explanation": explanation,
        "annotated_image": annotated_image
    }

# ============================================================
# ğŸ”Ÿ VISUALIZATION
# ============================================================

def visualize_results(image, diagnosis, recommendation):
    """
    Create visual output with annotations
    """
    annotated = image.copy()

    # Draw bounding box
    bbox = diagnosis['bbox']
    x1, y1, x2, y2 = map(int, bbox)

    # Color based on severity
    severity = diagnosis['severity']['severity']
    if severity == "High":
        color = (255, 0, 0)  # Red
    elif severity == "Moderate":
        color = (255, 165, 0)  # Orange
    else:
        color = (0, 255, 0)  # Green

    cv2.rectangle(annotated, (x1, y1), (x2, y2), color, 3)

    # Add label
    label = f"{diagnosis['crop_type']}: {diagnosis['disease_name']}"
    label2 = f"Severity: {severity} | Conf: {diagnosis['confidence']:.2f}"

    # Background for text
    cv2.rectangle(annotated, (x1, y1-60), (x2, y1), color, -1)
    cv2.putText(annotated, label, (x1+5, y1-35),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)
    cv2.putText(annotated, label2, (x1+5, y1-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

    return annotated

# ============================================================
# 1ï¸âƒ£1ï¸âƒ£ TEST THE PIPELINE
# ============================================================

# Test on validation images
test_image_dir = f'{dataset.location}/valid/images'
test_images = list(Path(test_image_dir).glob('*.jpg'))[:5]  # Test on 5 images

print("\n" + "="*60)
print("TESTING CROP HEALTH ANALYSIS PIPELINE")
print("="*60 + "\n")

for img_path in test_images:
    print(f"\nğŸ“· Analyzing: {img_path.name}")
    print("-" * 60)

    result = analyze_crop_image(str(img_path), model)

    if "error" in result:
        print(f"âŒ Error: {result['error']}")
        continue

    # Display results
    print(result['explanation'])

    # Show annotated image
    plt.figure(figsize=(12, 8))
    plt.imshow(result['annotated_image'])
    plt.axis('off')
    plt.title(f"{result['diagnosis']['crop_type']} - {result['diagnosis']['disease_name']}")
    plt.show()

    print("\n" + "="*60)

# ============================================================
# 1ï¸âƒ£2ï¸âƒ£ SAVE MODEL FOR DEPLOYMENT
# ============================================================

# Export to ONNX for edge deployment
model.export(format='onnx', dynamic=True, simplify=True)
print("\nâœ… Model exported to ONNX format for deployment!")

# Save complete pipeline configuration
pipeline_config = {
    "model_path": best_model_path,
    "confidence_threshold": 0.5,
    "crop_types": list(CROP_TYPE_MAPPING.values()),
    "disease_recommendations": DISEASE_RECOMMENDATIONS
}

with open('pipeline_config.json', 'w') as f:
    json.dump(pipeline_config, f, indent=2)

print("âœ… Pipeline configuration saved!")

# ============================================================
# 1ï¸âƒ£3ï¸âƒ£ INTERACTIVE TESTING FUNCTION
# ============================================================

def test_single_image(image_path):
    """
    Quick test function for single images
    """
    result = analyze_crop_image(image_path, model)

    if "error" not in result:
        print(result['explanation'])
        plt.figure(figsize=(12, 8))
        plt.imshow(result['annotated_image'])
        plt.axis('off')
        plt.show()

        return result
    else:
        print(f"Error: {result['error']}")
        return None

# Example usage:
# test_single_image('/path/to/your/image.jpg')

print("\n" + "="*60)
print("âœ… COMPLETE PIPELINE READY!")
print("="*60)
print("\nFeatures:")
print("âœ“ YOLOv8-based disease detection")
print("âœ“ Crop type identification")
print("âœ“ Disease severity estimation")
print("âœ“ Growth stage prediction")
print("âœ“ Rule-based recommendations")
print("âœ“ Farmer-friendly explanations")
print("âœ“ Visual annotations")
print("âœ“ ONNX export for edge deployment")
print("\nUse: test_single_image('your_image.jpg') to test!")
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

def visualize_crop_health(image_bgr, result):
    """
    Unified visualization for crop health analysis
    image_bgr: original image (BGR)
    result: output dict from analyze_crop_image()
    """

    diagnosis = result["diagnosis"]
    recommendation = result["recommendation"]

    # -------------------------------
    # 1ï¸âƒ£ Draw bounding box + label
    # -------------------------------
    img = image_bgr.copy()

    x1, y1, x2, y2 = map(int, diagnosis["bbox"])
    crop = diagnosis["crop_type"]
    disease = diagnosis["disease_name"]
    conf = diagnosis["confidence"]
    severity = diagnosis["severity"]["severity"]

    if severity == "High":
        color = (255, 0, 0)      # Red
    elif severity == "Moderate":
        color = (255, 165, 0)    # Orange
    else:
        color = (0, 200, 0)      # Green

    cv2.rectangle(img, (x1, y1), (x2, y2), color, 3)

    label = f"{crop} | {disease} | {severity} ({conf:.2f})"
    (tw, th), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 2)
    cv2.rectangle(img, (x1, y1 - th - 10), (x1 + tw + 6, y1), color, -1)
    cv2.putText(
        img, label, (x1 + 3, y1 - 5),
        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2
    )

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # -------------------------------
    # 2ï¸âƒ£ Create dashboard
    # -------------------------------
    plt.figure(figsize=(16, 9))

    # Image panel
    plt.subplot(2, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Detected Leaf & Disease")
    plt.axis("off")

    # -------------------------------
    # 3ï¸âƒ£ Severity visualization
    # -------------------------------
    severity_map = {"Low": 1, "Moderate": 2, "High": 3}
    sev_score = severity_map[severity]

    plt.subplot(2, 2, 2)
    plt.barh(
        ["Severity"],
        [sev_score],
        color=["green", "orange", "red"][sev_score - 1]
    )
    plt.xlim(0, 3)
    plt.xticks([1, 2, 3], ["Low", "Moderate", "High"])
    plt.title(f"Affected Area: {diagnosis['severity']['affected_percentage']}%")

    # -------------------------------
    # 4ï¸âƒ£ Growth stage visualization
    # -------------------------------
    growth = diagnosis["growth_stage"]

    plt.subplot(2, 2, 3)
    stages = ["Early/Seedling", "Vegetative", "Flowering/Mature"]
    stage_codes = [1, 2, 3]

    plt.barh(stages, stage_codes, color="lightgray")
    plt.barh(growth["stage"], growth["stage_code"], color="green")

    plt.xlim(0, 3.5)
    plt.title(f"Estimated Growth Stage (Confidence: {growth['confidence']})")

    # -------------------------------
    # 5ï¸âƒ£ Text summary
    # -------------------------------
    plt.subplot(2, 2, 4)
    plt.axis("off")

    summary_text = f"""
CROP: {crop}
DISEASE: {disease}
CONFIDENCE: {conf}

SEVERITY: {severity}
AFFECTED AREA: {diagnosis['severity']['affected_percentage']}%

GROWTH STAGE: {growth['stage']}

URGENCY: {recommendation['urgency']}
ACTION: {recommendation['action']}
TREATMENT: {recommendation['treatment']}
TIMING: {recommendation['timing']}
"""

    plt.text(0.01, 0.5, summary_text, fontsize=11, verticalalignment="center")
    plt.title("Diagnosis & Recommendation Summary")

    plt.tight_layout()
    plt.show()

```
